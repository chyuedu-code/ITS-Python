## 1. 自訂函式基礎
函式是將一段具體功能的程式碼封裝起來的區塊，可以重複呼叫並接收參數、回傳結果。

### 自建函式定義與呼叫語法：
```python
def 函式名稱(參數1, 參數2, ...):
    """函式說明文件字串 (選填)"""
    程式區塊
    return 回傳值1, 回傳值2, ...
```

> 💡 **知識小補充：**
> * 如果函式內部沒有寫 `return` 語法，Python 會在執行完畢後預設回傳 `None`。
> * Python 支援「多重回傳值」，如果回傳多個資料，實質上會被封裝成一個 **元組 (Tuple)** 回傳。

````python
# 基本自訂函式與多重回傳實作
def calculate_rectangle(width, height):
    area = width * height
    perimeter = (width + height) * 2
    return area, perimeter # 回傳多個值

# 呼叫函式並接收回傳值
rect_area, rect_peri = calculate_rectangle(5, 3)
print(f"矩形面積: {rect_area}, 矩形周長: {rect_peri}")

# 檢驗多重回傳值的真實型態
raw_result = calculate_rectangle(5, 3)
print("多重回傳未解包時的型態:", type(raw_result), raw_result)
````

## 2. 指定參數與參數預設值
在呼叫函式時，傳遞參數的機制非常靈活。主要有以下兩大特性：

### ① 指定參數 (Keyword Arguments)
呼叫時直接寫出 `參數名稱 = 數值`。一旦明確指定參數名稱，傳入的**順序就不再受限制**。

### ② 參數預設值 (Default Arguments)
定義函式時可以為參數賦予預設值（例如 `c=1`）。
* 呼叫時**若沒給**該參數，則自動採用預設值。
* 呼叫時**若有給**該參數，傳入的新變數會直接**取代（覆蓋）**預設值。

> ⚠️ **重要語法規則：** 在定義函式時，**有預設值的參數必須放在所有「沒有預設值參數」的後面**，否則會觸發 `SyntaxError`！

````python
# 1. 指定參數範例
def add_two(a, b):
    return a + b

res1 = add_two(a=3, b=5) # 明確指派
res2 = add_two(b=5, a=3) # 調換順序，結果相同
print(f"指定參數結果: res1={res1}, res2={res2}")

# 2. 參數預設值範例
def add_three(a, b, c=1):
    return a + b + c

res_default = add_three(3, 5)    # 未傳入 c，c 採用預設值 1 (3 + 5 + 1)
res_override = add_three(3, 5, 4) # 傳入 c=4，取代預設值 1 (3 + 5 + 4)
print(f"採用預設值: {res_default}")
print(f"取代預設值: {res_override}")
````

## 3. 模組匯入的三種用法 (Importing Modules)
模組（Module）是一個包含 Python 定義和語句的文件。以下是匯入模組最常用的三種語法特性對比：

### 用法 1：`import 模組`
* **特點**：最標準安全的方式。使用時必須加上模組名稱作為前綴字。
* **語法**：`模組.函式名稱(參數)`

### 用法 2：`import 模組 as 別名`
* **特點**：當模組名稱太長時，可以用 `as` 取一個簡短的別名，簡化代碼。
* **語法**：`別名.函式名稱(參數)`

### 用法 3：`from 模組 import *` 或 `from 模組 import 函式`
* **特點**：直接將模組內的函式抽離並匯入到當前的名稱空間。使用時**不需要**再加任何前綴字。
* **缺點**：容易與當前檔案自己寫的函式發生名稱衝突（不建議盲目使用 `import *`）。
* **語法**：`函式名稱(參數)`

````python
# 由於此環境無法創立多個外部實體 .py 檔案，這裡以內建模組 math 為例模擬三種用法

print("--- 用法 1：標準 import ---")
import math
print("math.sqrt(16) =", math.sqrt(16))

print("\n--- 用法 2：使用 as 別名 ---")
import math as m
print("m.pow(2, 3)   =", m.pow(2, 3))

print("\n--- 用法 3：from ... import ---")
from math import gcd
print("直接呼叫 gcd(12, 18) =", gcd(12, 18))
````
