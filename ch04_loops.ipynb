# Chapter 4: 迴圈 (Loops)
本章節將介紹 Python 中的計數迴圈 (`for`)、條件迴圈 (`while`)、數列生成工具 `range()`、巢狀迴圈，以及迴圈控制指令 `break` 與 `continue` 的用法。

## 1. range() 函式
`range()` 函式的主要功能是建立一組**整數等差數列**，常與 `for` 迴圈搭配使用。

### 三種常見的宣告方式：
1. **`range(終值)`**：從 `0` 開始，到 `終值 - 1` 結束（不包含終值）。
2. **`range(起始值, 終值)`**：從 `起始值` 開始，到 `終值 - 1` 結束。
3. **`range(起始值, 終值, 間隔值)`**：從 `起始值` 開始，每次累加 `間隔值`，直到不超過 `終值 - 1`。

````python
# range() 數列生成範例
print("range(5)        :", list(range(5)))
print("range(2, 7)     :", list(range(2, 7)))
print("range(1, 10, 2) :", list(range(1, 10, 2)))
print("range(10, 2, -2):", list(range(10, 2, -2)))  # 也可以遞減
````

## 2. for 迴圈 與 巢狀迴圈
用於巡覽走訪數列（或任何可迭代物件）中的每一個元素。

### 基本語法
```python
for 數 in 數列:
    程式區塊
```

### 巢狀迴圈 (Nested Loops)
當迴圈中又包含其他迴圈時，就形成了巢狀迴圈。最經典的應用就是印出九九乘法表。
```python
for i in XXX:
    for j in YYY:
        程式區塊
```

````python
# 基礎 for 迴圈範例
print("--- 基礎 for 迴圈 ---")
for i in range(1, 4):
    print(f"這是第 {i} 次執行")

# 巢狀迴圈範例：簡易乘法表 (2x2)
print("\n--- 巢狀迴圈範例 ---")
for i in range(1, 3):
    for j in range(1, 3):
        print(f"{i} x {j} = {i*j}")
````

## 3. while 迴圈
用於**沒有固定執行次數**的場景。只要條件式保持為 `True`，迴圈就會不斷執行，直到條件變為 `False` 為止。

### 基本語法
```python
while 條件式:
    程式區塊
```

````python
# while 迴圈範例：累加到不超過 10
count = 1
while count <= 4:
    print(f"while 迴圈 count 檔前值: {count}")
    count += 1
````

## 4. 迴圈控制指令：break 與 continue

* **`break`**：**徹底跳離**。當執行到 `break` 時，會強行中斷並「完全跳出」目前的整個迴圈，直接執行迴圈下方的其他程式碼。
* **`continue`**：**略過本次**。當執行到 `continue` 時，會「暫停目前這輪」後面尚未執行的程式，直接「跳回迴圈起始處」開始下一輪的判斷與執行。

[Image of loop control break vs continue flowchart]

````python
# break 實作範例
print("--- break 測試 (遇到 3 就跳出) ---")
for n in range(1, 6):
    if n == 3:
        print("  遇到 3 了！觸發 break，徹底離開迴圈。")
        break
    print(f"  數字: {n}")
print("迴圈結束後的程式繼續執行。")

# continue 實作範例
print("\n--- continue 測試 (跳過 3) ---")
for n in range(1, 6):
    if n == 3:
        print("  遇到 3 了！觸發 continue，跳過後面的列印。")
        continue
    print(f"  數字: {n}")
````

> 💡 **小提醒：** 
> 1. **冒號 `:` 與縮排**：不論是 `for`、`while` 還是內嵌的 `if`，句尾都別忘了加上冒號 `:`，且內部的程式區塊必須正確縮排。
> 2. **避免無窮迴圈 (Infinite Loop)**：撰寫 `while` 迴圈時，請務必確保迴圈內部有改變條件的機制（例如 `count += 1`）。如果條件永遠為 `True`，程式將陷入死循環鎖死。
