## 1. 檔案開啟與關閉

### 1.1 基本語法：`open()` 與 `close()`

在 Python 中，使用 `open()` 函數開啟檔案，使用完畢後必須用 `close()` 關閉檔案。

**基本語法：**
```python
file = open(filename, mode, encoding='utf-8')
# 進行檔案操作...
file.close()  # 必須手動關閉檔案
```

### 常用開啟模式 (Modes)：

| 模式 | 說明 | 檔案不存在 | 原內容 |
| :---: | :--- | :--- | :--- |
| `'r'` | 唯讀模式 (Read) | 報錯 | 保留 |
| `'w'` | 覆寫模式 (Write) | 自動建立 | **清空** |
| `'a'` | 續寫模式 (Append) | 自動建立 | 保留（從末尾寫入） |
| `'r+'` | 讀寫模式 | 報錯 | 保留 |
| `'w+'` | 讀寫模式 | 自動建立 | **清空** |
| `'a+'` | 讀寫續寫模式 | 自動建立 | 保留（從末尾寫入） |

> 💡 **提示：** 可加上 `'b'` 表示二進位模式，如 `'rb'`、`'wb'`，用於處理圖片、音訊等非文字檔案。

> ⚠️ **重要：** 使用 `open()` 後，務必記得呼叫 `close()`，否則可能造成：
> - 記憶體洩漏
> - 檔案鎖定（其他程式無法存取）
> - 資料未正確寫入檔案

````python
# 傳統方式：使用 open() 和 close()
print("=== 方法 1：傳統 open() 與 close() ===")

# 寫入檔案
file_path = "traditional_demo.txt"
f = open(file_path, "w", encoding="utf-8")
f.write("使用傳統方式寫入檔案\n")
f.write("記得要手動關閉檔案！")
f.close()  # 必須手動關閉
print(f"已寫入檔案：{file_path}")

# 讀取檔案
f = open(file_path, "r", encoding="utf-8")
content = f.read()
print(f"\n讀取內容：\n{content}")
f.close()  # 必須手動關閉

print("\n注意：如果忘記 close()，可能會有問題！")
````

### 1.2 推薦方式：`with` 語法（Context Manager）

Python 提供了更安全的 `with` 語法，它會**自動關閉檔案**，即使程式執行出錯也能確保檔案被正確關閉。

**語法：**
```python
with open(filename, mode, encoding='utf-8') as file:
    # 進行檔案操作...
    pass
# 離開 with 區塊後，檔案自動關閉
```

**優點：**
- ✅ 自動關閉檔案，無需手動呼叫 `close()`
- ✅ 即使發生錯誤也能確保檔案被關閉
- ✅ 程式碼更簡潔、安全
- ✅ **ITS Python 考試推薦寫法**

> 💡 **最佳實踐：** 在 ITS Python 考試和實際開發中，優先使用 `with` 語法！

````python
# 推薦方式：使用 with 語法
print("=== 方法 2：推薦使用 with 語法 ===")

file_path = "with_demo.txt"

# 寫入檔案（自動關閉）
with open(file_path, "w", encoding="utf-8") as f:
    f.write("使用 with 語法寫入檔案\n")
    f.write("離開 with 區塊後會自動關閉！")
print(f"已寫入檔案：{file_path}（已自動關閉）")

# 讀取檔案（自動關閉）
with open(file_path, "r", encoding="utf-8") as f:
    content = f.read()
    print(f"\n讀取內容：\n{content}")
print("\n檔案已自動關閉，無需手動 close()！")

# 驗證：離開 with 後，檔案已關閉
# print(f.read())  # 如果取消註解，會報錯：ValueError: I/O operation on closed file
````

### 1.3 兩種方式比較

| 比較項目 | `open()` + `close()` | `with open()` |
| :--- | :--- | :--- |
| 需要手動關閉 | ✅ 是（容易忘記） | ❌ 否（自動關閉） |
| 錯誤處理 | ⚠️ 出錯時可能未關閉 | ✅ 即使出錯也會關閉 |
| 程式碼簡潔度 | ⚠️ 較冗長 | ✅ 簡潔 |
| ITS 考試推薦 | ❌ | ✅ **強烈推薦** |

> 📌 **結論：** 除非有特殊需求，否則請優先使用 `with` 語法！

## 2. 基礎讀寫操作 (Read / Write)
* **`write(string)`**：將字串寫入檔案。
* **`read(n)`**：讀取整個檔案（若給予參數 `n`，則唯獨讀取前 `n` 個字元/位元組）。
* **`readline()`**：逐行讀取檔案。

````python
# 建立模擬檔案並寫入、讀取資料
file_path = "demo_ch9.txt"

print("--- 1. 寫入檔案 (Write) ---")
with open(file_path, "w", encoding="utf-8") as f:
    f.write("Python 程式設計\n")
    f.write("歡樂學習每一天")
print(f"已成功將內容寫入 {file_path} ")

print("\n--- 2. 讀取檔案 (Read) ---")
with open(file_path, "r", encoding="utf-8") as f:
    content = f.read()
    print("讀取到的全部內容:")
    print(content)
````

## 3. 進階指針控制：`tell()` 與 `seek()`
當您對檔案進行讀寫時，Python 內部有一個「檔案指針（File Pointer）」，就像打字時的游標一樣，記錄目前讀寫到了第幾個位置（單位：位元組 Byte）。

### 核心方法說明：
* **`f.tell()`**：**回傳**目前檔案指針（游標）所在的絕對位置。
* **`f.seek(offset, whence)`**：**移動**檔案指針到特定位置。
  * `offset`: 移動的位元組數（正數向後移動，負數向前移動）。
  * `whence`: 起始參考點 (預設為 0)。
    * `0`: 從檔案開頭計算。
    * `1`: 從目前指針位置計算（需在二進位模式 `'rb'` 下使用）。
    * `2`: 從檔案末尾計算（需在二進位模式 `'rb'` 下使用）。

> ⚠️ **編碼大陷阱：** 在 UTF-8 編碼中，一個英文字母或數字佔 **1 個位元組 (Byte)**，而一個常用中文字通常佔 **3 個位元組 (Bytes)**。因此在做 `seek()` 移動時，若切在中文的中間，讀取時會噴出 `UnicodeDecodeError`！

````python
# 指針 tell() 與 seek() 實作模擬
test_file = "pointer_test.txt"

# 寫入純英文數字以方便精準點算 Byte 位元組
with open(test_file, "w", encoding="utf-8") as f:
    f.write("ABCDE12345")

print("--- 檔案指針實驗 (內容: ABCDE12345) ---")
with open(test_file, "r", encoding="utf-8") as f:
    print(f"1. 剛開檔時，tell() 指針位置: {f.tell()}")
    
    # 讀取 3 個字元
    chunk = f.read(3)
    print(f"2. 讀取了 3 個字元: '{chunk}'")
    print(f"3. 讀完後，tell() 指針移到了位置: {f.tell()}")
    
    # 使用 seek() 將指針拉回開頭 (位置 0)
    f.seek(0)
    print(f"4. 執行 seek(0) 後，tell() 指針重置回: {f.tell()}")
    print(f"5. 再次讀取 2 個字元: '{f.read(2)}'")
    
    # 使用 seek() 跳過某些內容，例如移到位置 5
    f.seek(5)
    print(f"6. 執行 seek(5) 後，直接讀取剩餘內容: '{f.read()}'")
````
