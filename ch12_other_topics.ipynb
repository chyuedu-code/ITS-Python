## 1. `os` 模組：作業系統與檔案目錄操作
`os` 模組是 Python 與底層作業系統溝通的橋樑。舉凡檢查檔案、開關資料夾、刪除或建立路徑都離不開它。

### 💡 重點觀念提示：
* **`os.mkdir()`**：只能建立單層資料夾，如果父目錄不存在或資料夾已存在，會噴出 `FileExistsError`。
* **`os.makedirs()`**：可以一口氣建立**多層深層目錄**（例如 `data/img/today`），並可加上參數 `exist_ok=True` 來防止目錄已存在時報錯。
* **`os.path.join()`**：**跨平台開發必用！** Windows 系統使用反斜線 `\` 作為路徑分隔符，而 Linux/Mac 使用正斜線 `/`。此函式會自動根據執行環境補上正確的分隔符，避免路徑寫死造成程式崩潰。

````python
import os

print("--- 1. 當前路徑與環境檢查 ---")
current_dir = os.getcwd()
print("目前工作目錄 (cwd):", current_dir)

# 檢查特定名稱是否存在（此處檢查當前目錄）
print("當前工作目錄是否存在:", os.path.exists(current_dir))
print("這是一個資料夾嗎:", os.path.isdir(current_dir))

print("\n--- 2. 安全的路徑拼接 (跨平台最佳實踐) ---")
# 自動因應作業系統生成正確路徑 (Windows 為 data\img\a.png ; Mac/Linux 為 data/img/a.png)
safe_path = os.path.join("data", "img", "a.png")
print("安全路徑拼接結果:", safe_path)
print("提取純檔名 (basename):", os.path.basename(safe_path))
print("提取資料夾路徑 (dirname):", os.path.dirname(safe_path))

print("\n--- 3. 資料夾建立與動態環境整理 ---")
target_folder = "os_demo_folder"

if not os.path.exists(target_folder):
    os.mkdir(target_folder)
    print(f"成功建立單層資料夾: {target_folder}")
else:
    print(f"{target_folder} 已經存在，不重複建立。")

# 建立深層多層資料夾示範 (exist_ok=True 代表若存在不報錯)
os.makedirs(os.path.join(target_folder, "level1", "level2"), exist_ok=True)
print("成功建立多層巢狀資料夾！")

print("\n--- 4. 列出目錄內容 ---")
print("當前資料夾內的所有項目:", os.listdir("."))

# 環境清理清理：移除剛剛建的目錄（先清空子目錄才能刪除父目錄）
os.rmdir(os.path.join(target_folder, "level1", "level2"))
os.rmdir(os.path.join(target_folder, "level1"))
os.rmdir(target_folder)
print("\n[環境收拾] 示範目錄已成功清理完畢。")
````

## 2. `random` 模組：亂數與隨機應用全功能指南
在開發遊戲（如抽卡、洗牌）或是做數據隨機抽樣時，`random` 是最核心的工具。以下為其核心函式的參數與機制拆解：

| 函式名稱與參數 | 傳回值範圍 / 機制 | 特色與注意事項 |
| :--- | :--- | :--- |
| **`random.random()`** | `[0.0, 1.0)` 的浮點數 | 不需帶參數，包含 0 但**不包含 1**。 |
| **`random.randint(a, b)`** | `[a, b]` 的整數 | **包含 a 也包含 b** 的閉區間整數。 |
| **`random.randrange(start, stop, step)`** | 依據 range 規則的隨機整數 | **包含 start 但不包含 stop**。可指定步長（如只抽偶數）。 |
| **`random.choice(sequence)`** | 序列中的**單一個**元素 | 參數必須是可迭代序列（如串列、元組）。若序列為空會噴出 IndexError。 |
| **`random.sample(population, k)`** | 長度為 `k` 的新串列 | **不重複抽樣（不放回抽樣）**。`k` 不能大於群集總數。 |
| **`random.shuffle(sequence)`** | 無傳回值 (`None`) | **就地（In-place）打亂**原串列的順序。會直接改變原始資料。 |
| **`random.seed(a)`** | 無傳回值 | **亂數種子**。只要 seed 相同，每次產生的隨機序列就會完全一模一樣（用於科學再現）。 |

````python
import random

print("--- 1. 基礎數值隨機 ---")
print("random() [0.0 ~ 1.0 浮點數]:", random.random())
print("randint(1, 10) [包含 1 和 10 的整數]:", random.randint(1, 10))
print("randrange(10, 100, 10) [10到90之間的整十數]:", random.randrange(10, 100, 10))

print("\n--- 2. 序列與群集隨機操作 ---")
languages = ["Python", "Java", "C++", "JavaScript", "Go"]
print("原始串列:", languages)

print("choice() 隨機盲抽一個:", random.choice(languages))
print("sample(k=3) 隨機選3個不重複的項目:", random.sample(languages, k=3))

nested_list = languages.copy() # 複製一份來洗牌，避免破壞原始資料
random.shuffle(nested_list)
print("shuffle() 就地洗牌後的串列狀態:", nested_list)

print("\n--- 3. 亂數種子 (Seed) 的神奇魔法 ---")
random.seed(42)  # 設定固定種子為 42
print("種子固定時第一次抽籤 (randint):", random.randint(1, 1000))
random.seed(42)  # 重新灌入相同的種子
print("種子固定時第二次抽籤 (重新設定後):", random.randint(1, 1000))
````

## 3. `math` 模組：常用基礎數學函數
`math` 模組提供了豐富的數學運算函數。以下為最常用的基礎函數：

### 📊 常用函數對照表：

| 函式名稱 | 功能說明 | 範例 | 結果 |
| :--- | :--- | :--- | :--- |
| **`math.sqrt(x)`** | 平方根 | `math.sqrt(16)` | `4.0` |
| **`math.pow(x, y)`** | x 的 y 次方 | `math.pow(2, 3)` | `8.0` |
| **`math.ceil(x)`** | 無條件進位 | `math.ceil(4.2)` | `5` |
| **`math.floor(x)`** | 無條件捨去 | `math.floor(4.8)` | `4` |
| **`math.trunc(x)`** | 截斷小數部分 | `math.trunc(4.9)` | `4` |
| **`math.fabs(x)`** | 絕對值（浮點數） | `math.fabs(-5.5)` | `5.5` |
| **`math.factorial(x)`** | 階乘 (x!) | `math.factorial(5)` | `120` |
| **`math.gcd(a, b)`** | 最大公因數 | `math.gcd(48, 18)` | `6` |

### 🔢 常用數學常數：
| 常數名稱 | 說明 | 數值 |
| :--- | :--- | :--- |
| **`math.pi`** | 圓周率 π | `3.141592653589793` |
| **`math.e`** | 自然對數底數 e | `2.718281828459045` |

````python
import math

print("--- 1. 基礎數學運算 ---")
print("sqrt(16) 平方根:", math.sqrt(16))
print("pow(2, 3) 次方運算:", math.pow(2, 3))
print("factorial(5) 階乘:", math.factorial(5))
print("gcd(48, 18) 最大公因數:", math.gcd(48, 18))
print("fabs(-5.5) 絕對值:", math.fabs(-5.5))

print("\n--- 2. 取整函數 ---")
num = 4.567
print(f"原始數值: {num}")
print("ceil(4.567) 無條件進位:", math.ceil(num))
print("floor(4.567) 無條件捨去:", math.floor(num))
print("trunc(4.567) 截斷小數:", math.trunc(num))

print("\n--- 3. 數學常數 ---")
print("圓周率 π:", math.pi)
print("自然對數底數 e:", math.e)

print("\n--- 4. 實際應用範例：計算圓形面積與周長 ---")
radius = 5
area = math.pi * math.pow(radius, 2)
circumference = 2 * math.pi * radius
print(f"半徑 {radius} 的圓形：")
print(f"  面積 = {area:.2f}")
print(f"  周長 = {circumference:.2f}")

print("\n--- 5. 實際應用範例：計算直角三角形斜邊 (畢氏定理) ---")
a, b = 3, 4
c = math.sqrt(math.pow(a, 2) + math.pow(b, 2))
print(f"直角邊 a={a}, b={b}")
print(f"斜邊 c = {c}")
````

## 4. `csv` 模組：CSV 檔案的讀取與寫入
CSV（Comma-Separated Values，逗號分隔值）是一種常見的表格資料格式，廣泛用於資料交換。Python 的 `csv` 模組提供了簡便的方法來處理 CSV 檔案。

### 📊 核心功能：
* **`csv.reader()`**：讀取 CSV 檔案，每一行作為一個串列。
* **`csv.writer()`**：寫入 CSV 檔案，將資料寫入檔案中。
* **`csv.DictReader()`**：將 CSV 讀取為字典格式，欄位名稱作為鍵。
* **`csv.DictWriter()`**：將字典資料寫入 CSV 檔案。

````python
import csv

print("--- 1. 寫入 CSV 檔案 (使用 writer) ---")
# 準備資料
data = [
    ['姓名', '年齡', '城市'],
    ['Alice', '25', '台北'],
    ['Bob', '30', '高雄'],
    ['Charlie', '35', '台中']
]

# 寫入 CSV 檔案
with open('students.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(data)
print("CSV 檔案已成功寫入！")

print("\n--- 2. 讀取 CSV 檔案 (使用 reader) ---")
with open('students.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)

print("\n--- 3. 使用 DictWriter 寫入 (字典格式) ---")
dict_data = [
    {'姓名': 'David', '年齡': '28', '城市': '新竹'},
    {'姓名': 'Eve', '年齡': '32', '城市': '台南'}
]

with open('students_dict.csv', 'w', newline='', encoding='utf-8') as file:
    fieldnames = ['姓名', '年齡', '城市']
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()  # 寫入標題列
    writer.writerows(dict_data)
print("字典格式 CSV 檔案已寫入！")

print("\n--- 4. 使用 DictReader 讀取 (字典格式) ---")
with open('students_dict.csv', 'r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        print(f"姓名: {row['姓名']}, 年齡: {row['年齡']}, 城市: {row['城市']}")
````

## 5. `json` 模組：JSON 資料的讀取與寫入
JSON（JavaScript Object Notation）是一種輕量級的資料交換格式，易於閱讀和編寫。Python 的 `json` 模組讓我們能夠輕鬆處理 JSON 資料。

### 📝 核心方法：

| 方法 | 功能說明 | 用途 |
| :--- | :--- | :--- |
| **`json.dump(obj, file)`** | 將 Python 物件寫入檔案 | 儲存 JSON 到檔案 |
| **`json.dumps(obj)`** | 將 Python 物件轉為 JSON 字串 | 轉換為 JSON 字串 |
| **`json.load(file)`** | 從檔案讀取 JSON 並轉為 Python 物件 | 從檔案載入 JSON |
| **`json.loads(string)`** | 將 JSON 字串轉為 Python 物件 | 解析 JSON 字串 |

### 💡 資料型態對應：
* Python `dict` ↔ JSON `object`
* Python `list` ↔ JSON `array`
* Python `str` ↔ JSON `string`
* Python `int/float` ↔ JSON `number`
* Python `True/False` ↔ JSON `true/false`
* Python `None` ↔ JSON `null`

````python
import json

print("--- 1. Python 物件轉 JSON 字串 (dumps) ---")
data = {
    '姓名': 'Alice',
    '年齡': 25,
    '城市': '台北',
    '技能': ['Python', 'JavaScript', 'SQL'],
    '在職': True
}

json_string = json.dumps(data, ensure_ascii=False, indent=2)
print("JSON 字串：")
print(json_string)

print("\n--- 2. JSON 字串轉 Python 物件 (loads) ---")
parsed_data = json.loads(json_string)
print("解析後的 Python 字典：", parsed_data)
print("姓名：", parsed_data['姓名'])
print("技能：", parsed_data['技能'])

print("\n--- 3. 寫入 JSON 檔案 (dump) ---")
students = [
    {'姓名': 'Bob', '年齡': 30, '成績': {'數學': 85, '英文': 90}},
    {'姓名': 'Charlie', '年齡': 28, '成績': {'數學': 92, '英文': 88}}
]

with open('students.json', 'w', encoding='utf-8') as file:
    json.dump(students, file, ensure_ascii=False, indent=2)
print("JSON 檔案已成功寫入！")

print("\n--- 4. 從 JSON 檔案讀取 (load) ---")
with open('students.json', 'r', encoding='utf-8') as file:
    loaded_data = json.load(file)

print("從檔案載入的資料：")
for student in loaded_data:
    print(f"姓名: {student['姓名']}, 數學成績: {student['成績']['數學']}")

print("\n--- 5. 實際應用：處理 API 回應格式 ---")
api_response = '{"status": "success", "data": {"user": "Alice", "score": 95}}'
response_data = json.loads(api_response)
print(f"狀態: {response_data['status']}")
print(f"使用者: {response_data['data']['user']}")
print(f"分數: {response_data['data']['score']}")
````

## 6. Pydoc 與 Docstring：寫出 Python 標準說明書
在開發大型專案時，寫代碼更要寫說明文件。Python 內建的 `pydoc` 能自動讀取內部的 `Docstring`（用三引號括起來的說明文字），並將其編排成專業的文件規格。

### 📝 標準 Docstring 架構剖析：
一個符合標準規範的 Python 模組檔案，通常包含以下區塊：
1. **`DESCRIPTION`**：寫在檔案最頂端（任何代碼之前）的標準三引號文字，用來概述此模組的用途。
2. **`FUNCTIONS`**：函式下方的首行三引號字串，詳細說明功能、參數與回傳值。
3. **`DATA`**：模組中的全域變數或元數據（如系統預設全域屬性）。
4. **`AUTHOR` / `CREDITS`**：透過定義 `__author__` 與 `__copyright__` 等雙下底線變數來宣告版權與開發者資訊。

````python
"""
mymodule1 - 隨機與語言學習範例模組

DESCRIPTION:
本模組提供兩大基礎展示功能：
1. 取得 1 到 100 之間的隨機整數。
2. 隨機獲取一種熱門的程式語言。
"""

import random

# --- DATA / AUTHOR 宣告 ---
__author__ = "Jim Tester"
__copyright__ = "Copyright 2019, The Learning Project"
DEFAULT_LEVEL = "Advanced"

# --- FUNCTIONS 宣告 ---
def get_random():
    """returns a random number between 1 and 100"""
    return random.randint(1, 100)

def get_language():
    """returns a random programming language from the list"""
    languages = ["Python", "Go", "Kotlin"]
    return random.choice(languages)

print("--- 1. 在程式內動態查詢說明書 (help) ---")
# 我們可以直接對自定義的函式或內建的 math 模組呼叫 help()
help(get_random)
````

## 7. 在終端機 (Terminal / Command Line) 控制 pydoc
除了在程式內用 `help()`，我們更常在終端機利用命令列直接生成說明網頁文件：

### 🛠️ 四大黃金指令對照表：

#### 1. 終端機純文字預覽
```bash
python -m pydoc math
python -m pydoc datetime
```
* **意義**：以「模組（`-m`）」的形式去執行 Python 內建的 `pydoc`工具，將 `math` 或 `datetime` 說明文件直接以純文字格式噴在終端機視窗中（按 `q` 可退出）。

#### 2. 導出 HTML 文件檔案
```bash
python -m pydoc -w math
```
* **意義**：`-w` 代表 **Write**（寫入）。執行後，會在目前資料夾下自動產出一份名為 **`math.html`** 的離線網頁檔案，您可以直接雙擊用 Chrome 瀏覽器打開，介面極其精美專業！

#### 3. 啟動本機文件伺服器
```bash
python -m pydoc -b
```
* **意義**：`-b` 代表 **Browser**（瀏覽器）。會在您的電腦後台啟動一個 HTTP Server，並**自動幫您彈出瀏覽器網頁**，裡面包含了您本機安裝的所有 Python 套件的完整文件索引目錄！

#### 4. 自訂指定連接埠 (Port) 啟動
```bash
python -m pydoc -p 1234
```
* **意義**：`-p` 代表 **Port**（連接埠）。強制在指定通道（例如 `1234`）開啟文件伺服器，之後您可以手動在瀏覽器網址列輸入 `http://localhost:1234` 進入文件中心，適合用於遠端主機無圖形介面時的開發環境。
