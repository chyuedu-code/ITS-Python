# Chapter 10: 掌握 datetime 模組 (Date and Time Handling)
在程式開發中，時間的控制至關重要，舉凡**記錄系統日誌、計算程式執行時間、製作倒數計時器**，或是**產生存款交易的時間字串**等，都離不開 Python 內建的 `datetime` 模組。
本章將帶您深入學習四大核心類別（`date`、`time`、`datetime`、`timedelta`）與時間格式化的精準應用。

## 1. 核心類別簡介與匯入
在使用前，我們通常會從 `datetime` 模組中選取所需的類別進行匯入：
```python
from datetime import date, time, datetime, timedelta
```

### 核心四大類別分工：
1. **`date`**：純日期處理（年、月、日）。
2. **`time`**：純時間處理（時、分、秒、微秒）。
3. **`datetime`**：完整日期與時間結合（年、月、日、時、分、秒）。
4. **`timedelta`**：用來表示**「時間的長度/差值」**，是進行時間加減運算的核心。

## 2. `date` 與 `time` 類別的基礎操作
這兩個類別專門處理獨立的日期或時間資訊。

````python
from datetime import date, time

print("--- 1. date 類別示範 ---")
# 建立指定日期
d = date(2026, 2, 13)
print("自訂日期:", d) 
print("今天的日期:", date.today())
# 讀取特定屬性
print(f"拆解屬性 -> 年: {d.year}, 月: {d.month}, 日: {d.day}")

print("\n--- 2. time 類別示範 ---")
# 建立指定時間 (時, 分, 秒)
t = time(14, 30, 0)
print("自訂時間:", t)
# 讀取特定屬性
print(f"拆解屬性 -> 時: {t.hour}, 分: {t.minute}, 秒: {t.second}")
````

## 3. `datetime` 類別與時間格式化
`datetime` 是最常使用的類別，除了可以同時控制日期與時間，更包含了將時間物件轉換為字串的強大工具。

### 常用格式化代碼對照表：
| 格式代碼 | 實質意義 | 範例結果 |
| :---: | :--- | :--- |
| **`%Y`** | 4 位數完整年份 | `2026` |
| **`%y`** | 2 位數簡寫年份 | `26` |
| **`%m`** | 補零的 2 位數月份 | `02` |
| **`%B`** | 月份完整名稱 (英文) | `February` |
| **`%b`** | 月份縮寫名稱 (英文) | `Feb` |
| **`%d`** | 補零的 2 位數日期 | `13` |
| **`%H`** | 24 小時制時數 | `14` |
| **`%I`** | 12 小時制時數 | `02` |
| **`%M`** | 分鐘數 | `30` |
| **`%S`** | 秒數 | `00` |
| **`%p`** | 上午/下午標記 | `AM` / `PM` |
| **`%A`** | 星期完整名稱 (英文) | `Friday` |
| **`%a`** | 星期縮寫名稱 (英文) | `Fri` |

### 💡 進階招式：`strftime` 與 `strptime` 怎麼分？
* **`strftime()`**: 將時間物件 **轉化成 (Format)** 漂亮好讀的字串。 (Object ➔ String)
* **`strptime()`**: 將字串 **解析成 (Parse)** 時間物件，通常用於讀取外部資料。 (String ➔ Object)

````python
from datetime import datetime

print("--- 1. 取得當前完整時間 ---")
now = datetime.now()
print("現在系統時間:", now)

print("\n--- 2. 建立指定時間點 ---")
dt = datetime(2026, 2, 13, 14, 30, 0)
print("設定的時間點:", dt)
print(f"拆解屬性 -> {dt.year}年{dt.month}月{dt.day}日，{dt.hour}點{dt.minute}分")

print("\n--- 3. 時間物件轉字串 strftime() ---")
formatted_str = dt.strftime("%Y年%m月%d日 (%A) %I:%M %p")
print("格式化後的視覺化字串:", formatted_str)
````

## 4. `timedelta` 類別：時間的加減與差值計算
在 Python 中，兩個 `datetime` 物件可以直接相減，相減後的結果型態就是 `timedelta`。若要把某個時間點加上「一段特定長度的時間」，也必須使用 `timedelta`。

````python
from datetime import datetime, timedelta

now = datetime.now()
print("現在起點時間:", now)

print("\n--- 1. 時間的推算 (加與減) ---")
future = now + timedelta(days=7)    # 往後推算 7 天
past = now - timedelta(hours=3)     # 往前推算 3 小時

print("7 天後的未來:", future)
print("3 小時前的過去:", past)
print("未來與過去的總時間差 (future - past):", future - past)

print("\n--- 2. 計算精準天數差 ---")
d1 = datetime(2026, 2, 1)
d2 = datetime(2026, 2, 13)

diff = d2 - d1
print(f"從 {d1.date()} 到 {d2.date()} 一共相隔了: {diff.days} 天")
````
