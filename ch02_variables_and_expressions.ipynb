## 1. 算術運算子

| 運算子 | 說明 | 範例 | 結果 |
| :---: | :--- | :--- | :--- |
| `+` | 加法 | `a = 1 + 2` | `a = 3` |
| `-` | 減法 | `a = 20 - 12` | `a = 8` |
| `*` | 乘法 | `a = 2 * 3` | `a = 6` |
| `/` | 除法 | `a = 9 / 2` | `a = 4.5` |
| `//` | 除法取整數（無條件捨去） | `a = 9 // 2` | `a = 4` |
| `%` | 餘數 | `a = 9 % 2` | `a = 1` |
| `**` | 次方 | `a = 2 ** 3` | `a = 8` |

````python
# 算術運算子程式範例
print("--- 算術運算子範例 ---")
print("1 + 2 =", 1 + 2)
print("20 - 12 =", 20 - 12)
print("2 * 3 =", 2 * 3)
print("9 / 2 =", 9 / 2)
print("9 // 2 =", 9 // 2)
print("9 % 2 =", 9 % 2)
print("2 ** 3 =", 2 ** 3)
````

> 💡 **小提醒：** 與除法有關的運算式（如 `/`、`//`、`%`），請特別留心 **`ZeroDivisionError`**。當除數為 `0` 時，程式會發生錯誤並中斷執行。

````python
# ZeroDivisionError 錯誤捕捉範例
try:
    result = 9 / 0
except ZeroDivisionError as e:
    print("捕捉到錯誤:", e)
````

## 3. 比較運算子

| 運算子 | 說明 | 範例 |
| :---: | :--- | :--- |
| `>` | 大於（a 是否大於 b） | `a > b` |

| `>=` | 大於等於（a 是否大於等於 b） | `a >= b` |
| `<=` | 小於等於（a 是否小於等於 b） | `a <= b` |
| `==` | 等於（a 是否等於 b） | `a == b` |

````python
# 比較運算子程式範例
print("--- 比較運算子範例 ---")
a = 5
b = 3
print(f"當 a = {a}, b = {b} 時：")
print("a > b  :", a > b)
print("a < b  :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
print("a == b :", a == b)
print("a != b :", a != b)
````

### 3.1 鏈式比較（Chained Comparison）

Python 支援**鏈式比較**，可以在一個運算式中串接多個比較運算子，類似數學式的寫法。

**語法格式：**
```python
a < b < c  # 等同於 (a < b) and (b < c)
```

**常見用法：**

| 鏈式比較 | 等同於 | 說明 |
| :--- | :--- | :--- |
| `a < b < c` | `(a < b) and (b < c)` | 檢查 b 是否介於 a 和 c 之間 |
| `a <= b <= c` | `(a <= b) and (b <= c)` | 檢查 b 是否在範圍內（含邊界） |
| `a > b > c` | `(a > b) and (b > c)` | 檢查是否遞減順序 |
| `a == b == c` | `(a == b) and (b == c)` | 檢查三個值是否相等 |

> 💡 **優點：** 鏈式比較讓程式碼更簡潔易讀，且中間變數只會被評估一次。

````python
# 鏈式比較程式範例
print("--- 鏈式比較範例 ---")

# 範例 1：檢查數值是否在範圍內
x = 15
print(f"\n範例 1：檢查 {x} 是否介於 10 到 20 之間")
print(f"鏈式比較：10 < {x} < 20 =", 10 < x < 20)
print(f"傳統寫法：(10 < {x}) and ({x} < 20) =", (10 < x) and (x < 20))

# 範例 2：包含邊界的範圍檢查（常用於分數判斷）
score = 85
print(f"\n範例 2：檢查分數 {score} 是否在 60-100 分之間")
print(f"60 <= {score} <= 100 =", 60 <= score <= 100)

# 範例 3：多個比較運算子組合
a, b, c = 10, 5, 2
print(f"\n範例 3：檢查 {a}, {b}, {c} 是否為遞減順序")
print(f"{a} > {b} > {c} =", a > b > c)
````

## 4. 邏輯運算子

邏輯運算子用於組合多個條件判斷，返回布林值（`True` 或 `False`）。

| 運算子 | 說明 | 範例 | 結果 |
| :---: | :--- | :--- | :--- |
| `and` | 邏輯且（兩個條件都為 True 時才返回 True） | `True and False` | `False` |
| `or` | 邏輯或（至少一個條件為 True 時返回 True） | `True or False` | `True` |
| `not` | 邏輯非（反轉布林值） | `not True` | `False` |

### 真值表（Truth Table）

**`and` 運算子：**

| A | B | A and B |
| :---: | :---: | :---: |
| True | True | True |
| True | False | False |
| False | True | False |
| False | False | False |

**`or` 運算子：**

| A | B | A or B |
| :---: | :---: | :---: |
| True | True | True |
| True | False | True |
| False | True | True |
| False | False | False |

**`not` 運算子：**

| A | not A |
| :---: | :---: |
| True | False |
| False | True |

````python
# 邏輯運算子基礎範例
print("--- 邏輯運算子基礎範例 ---")

# and 運算子
print("\n【and 運算子】")
print("True and True  :", True and True)
print("True and False :", True and False)
print("False and True :", False and True)
print("False and False:", False and False)

# or 運算子
print("\n【or 運算子】")
print("True or True   :", True or True)
print("True or False  :", True or False)
print("False or True  :", False or True)
print("False or False :", False or False)

# not 運算子
print("\n【not 運算子】")
print("not True       :", not True)
print("not False      :", not False)
````

````python
# 邏輯運算子實際應用範例
print("--- 邏輯運算子實際應用範例 ---")

# 範例 1：使用 and 檢查多個條件
age = 25
has_license = True
print(f"\n範例 1：檢查是否可以租車（年齡 >= 18 且有駕照）")
print(f"年齡：{age}，有駕照：{has_license}")
can_rent = age >= 18 and has_license
print(f"可以租車：{can_rent}")

# 範例 2：使用 or 檢查多個條件
is_weekend = True
is_holiday = False
print(f"\n範例 2：檢查是否可以休息（週末或假日）")
print(f"是週末：{is_weekend}，是假日：{is_holiday}")
can_rest = is_weekend or is_holiday
print(f"可以休息：{can_rest}")

# 範例 3：使用 not 反轉條件
is_raining = False
print(f"\n範例 3：檢查是否適合出門（沒下雨）")
print(f"正在下雨：{is_raining}")
can_go_out = not is_raining
print(f"適合出門：{can_go_out}")

# 範例 4：組合多個邏輯運算子
score = 85
attendance = 90
print(f"\n範例 4：檢查是否及格（分數 >= 60 且出席率 >= 80）")
print(f"分數：{score}，出席率：{attendance}%")
is_pass = score >= 60 and attendance >= 80
print(f"是否及格：{is_pass}")

# 範例 5：複雜條件判斷
temperature = 28
is_sunny = True
print(f"\n範例 5：檢查是否適合游泳（溫度 >= 25 且天氣晴朗）")
print(f"溫度：{temperature}°C，天氣晴朗：{is_sunny}")
good_for_swimming = temperature >= 25 and is_sunny
print(f"適合游泳：{good_for_swimming}")
````

### 4.1 短路求值（Short-circuit Evaluation）

Python 的邏輯運算子具有**短路求值**特性：

- **`and` 運算子**：如果左邊的條件為 `False`，右邊的條件不會被評估（因為結果必定為 `False`）
- **`or` 運算子**：如果左邊的條件為 `True`，右邊的條件不會被評估（因為結果必定為 `True`）

> 💡 **優點：** 短路求值可以提升程式效能，並避免不必要的運算或錯誤。

### 4.2 運算子優先順序

當多個運算子組合使用時，優先順序由高到低為：

1. `not`（最高優先）
2. `and`
3. `or`（最低優先）

建議使用括號 `()` 來明確表達運算順序，提高程式可讀性。

````python
# 短路求值與優先順序範例
print("--- 短路求值與優先順序範例 ---")

# 範例 1：and 的短路求值
print("\n範例 1：and 的短路求值")
x = 5
result = (x > 10) and (print("這行不會被執行") or True)
print(f"結果：{result}")
print("說明：因為 x > 10 為 False，右側運算式不會被評估")

# 範例 2：or 的短路求值
print("\n範例 2：or 的短路求值")
y = 15
result = (y > 10) or (print("這行不會被執行") or False)
print(f"結果：{result}")
print("說明：因為 y > 10 為 True，右側運算式不會被評估")

# 範例 3：運算子優先順序
print("\n範例 3：運算子優先順序")
a, b, c = True, False, True
print(f"a = {a}, b = {b}, c = {c}")
print(f"a or b and c = {a or b and c}")
print("說明：等同於 a or (b and c)，因為 and 優先於 or")
print(f"(a or b) and c = {(a or b) and c}")
print("說明：使用括號改變運算順序")

# 範例 4：not 的優先順序
print("\n範例 4：not 的優先順序")
x = True
y = False
print(f"x = {x}, y = {y}")
print(f"not x and y = {not x and y}")
print("說明：等同於 (not x) and y，因為 not 優先於 and")
print(f"not (x and y) = {not (x and y)}")
print("說明：使用括號改變運算順序")

# 範例 5：實際應用 - 複雜條件判斷
print("\n範例 5：實際應用 - 會員優惠檢查")
is_member = True
purchase_amount = 1500
has_coupon = False
print(f"是會員：{is_member}，消費金額：{purchase_amount}，有優惠券：{has_coupon}")

# 會員且消費超過1000，或有優惠券
get_discount = (is_member and purchase_amount > 1000) or has_coupon
print(f"可獲得折扣：{get_discount}")
print("條件：(會員 且 消費>1000) 或 有優惠券")
````

## 5. 變數的作用域（Scope）與生命週期（Lifetime）

變數的**作用域（Scope）**決定了變數可以在程式的哪些部分被存取，而**生命週期（Lifetime）**則決定了變數存在的時間長度。

### 📌 作用域的種類：

Python 遵循 **LEGB 規則**來查找變數：

1. **L (Local)** - 區域作用域：函式內部定義的變數
2. **E (Enclosing)** - 封閉作用域：巢狀函式的外層函式
3. **G (Global)** - 全域作用域：模組層級定義的變數
4. **B (Built-in)** - 內建作用域：Python 內建的名稱（如 `print`, `len` 等）

### 💡 核心概念：

| 變數類型 | 定義位置 | 作用域 | 生命週期 |
| :--- | :--- | :--- | :--- |
| **全域變數 (Global)** | 函式外部 | 整個模組 | 程式執行期間 |
| **區域變數 (Local)** | 函式內部 | 僅在函式內 | 函式執行期間 |

### 🔑 關鍵字：

* **`global`**：在函式內宣告要使用全域變數
* **`nonlocal`**：在巢狀函式內宣告要使用外層函式的變數

````python
print("--- 1. 全域變數 vs 區域變數 ---")

# 全域變數
global_var = "我是全域變數"

def example_function():
    # 區域變數
    local_var = "我是區域變數"
    print("函式內部：")
    print(f"  可以存取全域變數: {global_var}")
    print(f"  可以存取區域變數: {local_var}")

example_function()

print("\n函式外部：")
print(f"  可以存取全域變數: {global_var}")
# print(f"  可以存取區域變數: {local_var}")  # 這行會報錯！區域變數在函式外不存在
print("  無法存取區域變數（會產生 NameError）")
````

````python
print("--- 2. 變數名稱遮蔽（Variable Shadowing）---")

x = 100  # 全域變數

def shadowing_example():
    x = 10  # 區域變數，會「遮蔽」全域變數
    print(f"函式內的 x: {x}")

print(f"函式外的 x: {x}")
shadowing_example()
print(f"函式執行後，函式外的 x 仍然是: {x}")
print("\n說明：函式內的 x 是獨立的區域變數，不影響全域變數")
````

````python
print("--- 3. 使用 global 關鍵字修改全域變數 ---")

counter = 0  # 全域變數

def increment_wrong():
    # 錯誤示範：嘗試修改全域變數但沒有宣告 global
    # counter = counter + 1  # 這行會報錯！UnboundLocalError
    pass

def increment_correct():
    global counter  # 宣告要使用全域變數
    counter = counter + 1
    print(f"  函式內：counter = {counter}")

print(f"初始值：counter = {counter}")

increment_correct()
print(f"第一次呼叫後：counter = {counter}")

increment_correct()
print(f"第二次呼叫後：counter = {counter}")

print("\n說明：使用 global 關鍵字，函式才能修改全域變數")
````

````python
print("--- 4. LEGB 規則示範 ---")

x = "全域 (Global)"

def outer():
    x = "外層函式 (Enclosing)"
    
    def inner():
        x = "內層函式 (Local)"
        print(f"inner() 中的 x: {x}")
    
    inner()
    print(f"outer() 中的 x: {x}")

outer()
print(f"全域中的 x: {x}")

print("\n說明：Python 會優先使用最近作用域的變數 (Local > Enclosing > Global)")
````

````python
print("--- 5. 使用 nonlocal 關鍵字 ---")

def outer_function():
    count = 0  # 外層函式的變數
    
    def increment():
        nonlocal count  # 宣告要使用外層函式的變數
        count += 1
        print(f"  increment() 執行後：count = {count}")
    
    print(f"初始值：count = {count}")
    increment()
    increment()
    print(f"最終值：count = {count}")

outer_function()
print("\n說明：nonlocal 讓內層函式可以修改外層函式的變數")
````

````python
print("--- 6. 變數生命週期示範 ---")

def lifecycle_demo():
    temp = "我只在函式執行期間存在"
    print(f"函式內：{temp}")
    # 函式結束後，temp 變數會被銷毀

print("第一次呼叫函式：")
lifecycle_demo()

print("\n第二次呼叫函式：")
lifecycle_demo()

print("\n說明：每次呼叫函式，區域變數都會重新建立，函式結束後就被銷毀")
````

````python
print("--- 7. 實際應用：計數器實作 ---")

# 方法 1：使用全域變數（不推薦）
global_counter = 0

def count_global():
    global global_counter
    global_counter += 1
    return global_counter

print("\n使用全域變數：")
print(f"第 1 次：{count_global()}")
print(f"第 2 次：{count_global()}")
print(f"第 3 次：{count_global()}")

# 方法 2：使用閉包（推薦）
def create_counter():
    count = 0
    
    def increment():
        nonlocal count
        count += 1
        return count
    
    return increment

counter1 = create_counter()
counter2 = create_counter()

print("\n使用閉包（每個計數器獨立）：")
print(f"counter1 第 1 次：{counter1()}")
print(f"counter1 第 2 次：{counter1()}")
print(f"counter2 第 1 次：{counter2()}")
print(f"counter1 第 3 次：{counter1()}")
print(f"counter2 第 2 次：{counter2()}")

print("\n說明：閉包方式可以建立多個獨立的計數器，更靈活且安全")
````

### 🎯 最佳實踐建議：

1. **盡量避免使用全域變數**：全域變數容易造成程式難以維護和除錯
2. **優先使用函式參數和回傳值**：讓函式的輸入輸出更明確
3. **必要時才使用 `global`**：通常只在設定檔或常數時使用
4. **善用閉包**：比全域變數更安全、更靈活
5. **變數命名要清晰**：避免區域變數和全域變數同名造成混淆
