本章節將介紹 Python 中的基礎算術運算子、複合賦值運算子以及比較運算子，並提供相關範例與注意事項。

## 1. 算術運算子

| 運算子 | 說明 | 範例 | 結果 |
| :---: | :--- | :--- | :--- |
| `+` | 加法 | `a = 1 + 2` | `a = 3` |
| `-` | 減法 | `a = 20 - 12` | `a = 8` |
| `*` | 乘法 | `a = 2 * 3` | `a = 6` |
| `/` | 除法 | `a = 9 / 2` | `a = 4.5` |
| `//` | 除法取整數（無條件捨去） | `a = 9 // 2` | `a = 4` |
| `%` | 餘數 | `a = 9 % 2` | `a = 1` |
| `**` | 次方 | `a = 2 ** 3` | `a = 8` |

````python
# 算術運算子程式範例
print("--- 算術運算子範例 ---")
print("1 + 2 =", 1 + 2)
print("20 - 12 =", 20 - 12)
print("2 * 3 =", 2 * 3)
print("9 / 2 =", 9 / 2)
print("9 // 2 =", 9 // 2)
print("9 % 2 =", 9 % 2)
print("2 ** 3 =", 2 ** 3)
````

> 💡 **小提醒：** 與除法有關的運算式（如 `/`、`//`、`%`），請特別留心 **`ZeroDivisionError`**。當除數為 `0` 時，程式會發生錯誤並中斷執行。

````python
# ZeroDivisionError 錯誤捕捉範例
try:
    result = 9 / 0
except ZeroDivisionError as e:
    print("捕捉到錯誤:", e)
````

## 2. 算術 + 賦值運算子 (複合賦值運算子)

| 運算子 | 範例 | 等同於 |
| :---: | :--- | :--- |
| `+=` | `a += 1` | `a = a + 1` |
| `-=` | `a -= 1` | `a = a - 1` |
| `*=` | `a *= 2` | `a = a * 2` |
| `/=` | `a /= 2` | `a = a / 2` |
| `//=` | `a //= 3` | `a = a // 3` |
| `%=` | `a %= 3` | `a = a % 3` |
| `**=` | `a **= 2` | `a = a ** 2` |

````python
# 算術 + 賦值運算子程式範例
print("--- 複合賦值運算子範例 ---")
a = 10
a += 1
print("a += 1 ->", a)

a = 10
a -= 1
print("a -= 1 ->", a)

a = 10
a *= 2
print("a *= 2 ->", a)

a = 10
a /= 2
print("a /= 2 ->", a)

a = 10
a //= 3
print("a //= 3 ->", a)

a = 10
a %= 3
print("a %= 3 ->", a)

a = 10
a **= 2
print("a **= 2 ->", a)
````

## 3. 比較運算子

| 運算子 | 說明 | 範例 |
| :---: | :--- | :--- |
| `>` | 大於（a 是否大於 b） | `a > b` |
| `<` | 小於（a 是否小於 b） | `a < b` |
| `>=` | 大於等於（a 是否大於等於 b） | `a >= b` |
| `<=` | 小於等於（a 是否小於等於 b） | `a <= b` |
| `==` | 等於（a 是否等於 b） | `a == b` |
| `!=` | 不等於（a 是否不等於 b） | `a != b` |

````python
# 比較運算子程式範例
print("--- 比較運算子範例 ---")
a = 5
b = 3
print(f"當 a = {a}, b = {b} 時：")
print("a > b  :", a > b)
print("a < b  :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
print("a == b :", a == b)
print("a != b :", a != b)
````

### 3.1 鏈式比較（Chained Comparison）

Python 支援**鏈式比較**，可以在一個運算式中串接多個比較運算子，類似數學式的寫法。

**語法格式：**
```python
a < b < c  # 等同於 (a < b) and (b < c)
```

**常見用法：**

| 鏈式比較 | 等同於 | 說明 |
| :--- | :--- | :--- |
| `a < b < c` | `(a < b) and (b < c)` | 檢查 b 是否介於 a 和 c 之間 |
| `a <= b <= c` | `(a <= b) and (b <= c)` | 檢查 b 是否在範圍內（含邊界） |
| `a > b > c` | `(a > b) and (b > c)` | 檢查是否遞減順序 |
| `a == b == c` | `(a == b) and (b == c)` | 檢查三個值是否相等 |

> 💡 **優點：** 鏈式比較讓程式碼更簡潔易讀，且中間變數只會被評估一次。

````python
# 鏈式比較程式範例
print("--- 鏈式比較範例 ---")

# 範例 1：檢查數值是否在範圍內
x = 15
print(f"\n範例 1：檢查 {x} 是否介於 10 到 20 之間")
print(f"鏈式比較：10 < {x} < 20 =", 10 < x < 20)
print(f"傳統寫法：(10 < {x}) and ({x} < 20) =", (10 < x) and (x < 20))

# 範例 2：包含邊界的範圍檢查（常用於分數判斷）
score = 85
print(f"\n範例 2：檢查分數 {score} 是否在 60-100 分之間")
print(f"60 <= {score} <= 100 =", 60 <= score <= 100)

# 範例 3：多個比較運算子組合
a, b, c = 10, 5, 2
print(f"\n範例 3：檢查 {a}, {b}, {c} 是否為遞減順序")
print(f"{a} > {b} > {c} =", a > b > c)
````

## 4. 邏輯運算子

邏輯運算子用於組合多個條件判斷，返回布林值（`True` 或 `False`）。

| 運算子 | 說明 | 範例 | 結果 |
| :---: | :--- | :--- | :--- |
| `and` | 邏輯且（兩個條件都為 True 時才返回 True） | `True and False` | `False` |
| `or` | 邏輯或（至少一個條件為 True 時返回 True） | `True or False` | `True` |
| `not` | 邏輯非（反轉布林值） | `not True` | `False` |

### 真值表（Truth Table）

**`and` 運算子：**

| A | B | A and B |
| :---: | :---: | :---: |
| True | True | True |
| True | False | False |
| False | True | False |
| False | False | False |

**`or` 運算子：**

| A | B | A or B |
| :---: | :---: | :---: |
| True | True | True |
| True | False | True |
| False | True | True |
| False | False | False |

**`not` 運算子：**

| A | not A |
| :---: | :---: |
| True | False |
| False | True |

````python
# 邏輯運算子基礎範例
print("--- 邏輯運算子基礎範例 ---")

# and 運算子
print("\n【and 運算子】")
print("True and True  :", True and True)
print("True and False :", True and False)
print("False and True :", False and True)
print("False and False:", False and False)

# or 運算子
print("\n【or 運算子】")
print("True or True   :", True or True)
print("True or False  :", True or False)
print("False or True  :", False or True)
print("False or False :", False or False)

# not 運算子
print("\n【not 運算子】")
print("not True       :", not True)
print("not False      :", not False)
````

````python
# 邏輯運算子實際應用範例
print("--- 邏輯運算子實際應用範例 ---")

# 範例 1：使用 and 檢查多個條件
age = 25
has_license = True
print(f"\n範例 1：檢查是否可以租車（年齡 >= 18 且有駕照）")
print(f"年齡：{age}，有駕照：{has_license}")
can_rent = age >= 18 and has_license
print(f"可以租車：{can_rent}")

# 範例 2：使用 or 檢查多個條件
is_weekend = True
is_holiday = False
print(f"\n範例 2：檢查是否可以休息（週末或假日）")
print(f"是週末：{is_weekend}，是假日：{is_holiday}")
can_rest = is_weekend or is_holiday
print(f"可以休息：{can_rest}")

# 範例 3：使用 not 反轉條件
is_raining = False
print(f"\n範例 3：檢查是否適合出門（沒下雨）")
print(f"正在下雨：{is_raining}")
can_go_out = not is_raining
print(f"適合出門：{can_go_out}")

# 範例 4：組合多個邏輯運算子
score = 85
attendance = 90
print(f"\n範例 4：檢查是否及格（分數 >= 60 且出席率 >= 80）")
print(f"分數：{score}，出席率：{attendance}%")
is_pass = score >= 60 and attendance >= 80
print(f"是否及格：{is_pass}")

# 範例 5：複雜條件判斷
temperature = 28
is_sunny = True
print(f"\n範例 5：檢查是否適合游泳（溫度 >= 25 且天氣晴朗）")
print(f"溫度：{temperature}°C，天氣晴朗：{is_sunny}")
good_for_swimming = temperature >= 25 and is_sunny
print(f"適合游泳：{good_for_swimming}")
````

### 4.1 短路求值（Short-circuit Evaluation）

Python 的邏輯運算子具有**短路求值**特性：

- **`and` 運算子**：如果左邊的條件為 `False`，右邊的條件不會被評估（因為結果必定為 `False`）
- **`or` 運算子**：如果左邊的條件為 `True`，右邊的條件不會被評估（因為結果必定為 `True`）

> 💡 **優點：** 短路求值可以提升程式效能，並避免不必要的運算或錯誤。

### 4.2 運算子優先順序

當多個運算子組合使用時，優先順序由高到低為：

1. `not`（最高優先）
2. `and`
3. `or`（最低優先）

建議使用括號 `()` 來明確表達運算順序，提高程式可讀性。

````python
# 短路求值與優先順序範例
print("--- 短路求值與優先順序範例 ---")

# 範例 1：and 的短路求值
print("\n範例 1：and 的短路求值")
x = 5
result = (x > 10) and (print("這行不會被執行") or True)
print(f"結果：{result}")
print("說明：因為 x > 10 為 False，右側運算式不會被評估")

# 範例 2：or 的短路求值
print("\n範例 2：or 的短路求值")
y = 15
result = (y > 10) or (print("這行不會被執行") or False)
print(f"結果：{result}")
print("說明：因為 y > 10 為 True，右側運算式不會被評估")

# 範例 3：運算子優先順序
print("\n範例 3：運算子優先順序")
a, b, c = True, False, True
print(f"a = {a}, b = {b}, c = {c}")
print(f"a or b and c = {a or b and c}")
print("說明：等同於 a or (b and c)，因為 and 優先於 or")
print(f"(a or b) and c = {(a or b) and c}")
print("說明：使用括號改變運算順序")

# 範例 4：not 的優先順序
print("\n範例 4：not 的優先順序")
x = True
y = False
print(f"x = {x}, y = {y}")
print(f"not x and y = {not x and y}")
print("說明：等同於 (not x) and y，因為 not 優先於 and")
print(f"not (x and y) = {not (x and y)}")
print("說明：使用括號改變運算順序")

# 範例 5：實際應用 - 複雜條件判斷
print("\n範例 5：實際應用 - 會員優惠檢查")
is_member = True
purchase_amount = 1500
has_coupon = False
print(f"是會員：{is_member}，消費金額：{purchase_amount}，有優惠券：{has_coupon}")

# 會員且消費超過1000，或有優惠券
get_discount = (is_member and purchase_amount > 1000) or has_coupon
print(f"可獲得折扣：{get_discount}")
print("條件：(會員 且 消費>1000) 或 有優惠券")
````
