## 1. 基礎變數型態 (Data Types)
Python 是一種不需要事先宣告型態的動態語言，當你把值賦予給變數時，Python 會自動判斷其型態。

* **`int` (Integer, 整數)**：不帶有小數點的完整數字，包含正整數、負整數與零。
* **`float` (Floating-point, 浮點數)**：帶有小數點的數字，在科學計算或精準度要求高的場景必用。
* **`str` (String, 字串)**：用來表示文字資料，必須使用引號包覆起來。

```python
# 變數宣告與賦值
age = 25              # 整數 (int)
price = 199.95        # 浮點數 (float)
name = "Victor"       # 字串 (str)

print("--- 1. 變數數值輸出 ---")
print("年齡:", age)
print("價格:", price)
print("姓名:", name)
```

## 2. print() 函數的進階參數：`sep` 與 `end`
`print()` 是 Python 最常用的輸出函數，除了基本的輸出功能外，它還提供兩個非常實用的參數來控制輸出格式：

### 📌 2.1 `sep` (separator, 分隔符號)
* **功能**：當 `print()` 接收多個參數時，預設會用**空格**分隔它們。透過 `sep` 參數，你可以自訂分隔符號。
* **預設值**：`sep=' '` (一個空格)
* **常見應用**：輸出 CSV 格式、自訂分隔符號、無分隔輸出等。

### 📌 2.2 `end` (結尾字元)
* **功能**：`print()` 執行完畢後，預設會**自動換行**（加上 `\n`）。透過 `end` 參數，你可以改變結尾字元。
* **預設值**：`end='\n'` (換行符號)
* **常見應用**：同一行連續輸出、進度條顯示、自訂結尾符號等。

```python
print("--- 2.1 sep 參數示範 ---")
# 預設情況：用空格分隔
print("蘋果", "香蕉", "橘子")

# 使用逗號分隔
print("蘋果", "香蕉", "橘子", sep=", ")

# 使用 | 符號分隔
print("蘋果", "香蕉", "橘子", sep=" | ")

# 無分隔符號（直接連接）
print("2026", "06", "19", sep="")

# 使用 - 分隔日期
print("2026", "06", "19", sep="-")

print("\n--- 2.2 end 參數示範 ---")
# 預設情況：自動換行
print("第一行")
print("第二行")

# 使用空字串結尾（不換行）
print("\n使用 end='' 不換行：")
print("Hello", end="")
print("World")

# 使用空格結尾（同一行輸出）
print("\n使用 end=' ' 同行輸出：")
print("Python", end=" ")
print("is", end=" ")
print("awesome!")

# 自訂結尾符號
print("\n使用自訂結尾符號：")
print("項目 1", end=" >>> ")
print("項目 2", end=" >>> ")
print("項目 3")

print("\n--- 2.3 sep 與 end 組合應用 ---")
# 製作簡單的進度顯示
print("載入中", end="")
print(".", ".", ".", sep="", end=" ")
print("完成！")

# 輸出表格式資料
print("\n姓名", "年齡", "城市", sep="\t")
print("Alice", "25", "台北", sep="\t")
print("Bob", "30", "高雄", sep="\t")
```

## 3. 字串的宣告與魔法三引號
在 Python 中表示字串非常彈性，主要有以下三種包覆方式：

* **單引號 (`'...'`) 與雙引號 (`"..."`)**：在 Python 中**完全等價**。最棒的應用是當字串內部有引號時，可以外雙內單、或外單內雙互相嵌套，免去程式報錯。
* **連續三個雙引號 (`"""..."""`)**：**多行字串神器**。一般字串內如果直接按 Enter 換行會語法錯誤，但三引號字串允許你自由地打出大段多行文字，且裡面的換行格式會原封不動被保留下來！

```python
print("--- 3.1 單雙引號的靈活嵌套 ---")
# 外層雙引號，內層就可以安全放單引號
quote1 = "Python is famous for its 'simple' philosophy."
# 外層單引號，內層就可以放雙引號
quote2 = '媽媽大喊: "快點來吃晚餐！"'

print(quote1)
print(quote2)

print("\n--- 3.2 連續三個雙引號的多行文字 ---")
multi_line_str = """親愛的學員您好：
恭喜您成功加入了 Python 自動化開發課程。
祝您學習愉快！
              -- 導師團隊 敬上"""

print(multi_line_str)
```

## 4. 跳脫字元 (Escape Characters)：`\n` 與 `\t`
當我們想在單行或普通的雙引號字串中，加入「換行」或「大空格（Tab鍵）」時，就必須使用反斜線 `\` 開頭的跳脫字元：

* **`\n` (Newline)**：代表**換行**。程式讀到它時，後面的文字會直接下推到下一行。
* **`\t` (Tab)**：代表**水平定位跳格**。相當於在鍵盤上按下 Tab 鍵，會留出一段標準的空白縮排，非常適合用來對齊排版文字。

```python
print("--- 4.1 換行字元 \\n 範例 ---")
lines = "第一行文字\n第二行文字\n第三行文字"
print(lines)

print("\n--- 4.2 縮排字元 \\t 欄位對齊範例 ---")
header = "品名\t\t單價\t數量"
row1 = "蘋果(Apple)\t$30\t5"
row2 = "香蕉(Banana)\t$15\t12"

print(header)
print(row1)
print(row2)
```

## 5. 註解 (Comments)
註解是寫給人類工程師看的備忘錄，Python 解譯器執行到這些區塊時會**完全忽略**。它能大幅提升程式碼的可讀性。

* **單行註解 (`#`)**：在井字號後面寫的所有文字，在該行之內都不會被執行。
* **多行註解**：在實務上，我們常利用「未賦值的三引號字串 `"""`」放在程式區塊中，當作大段的多行註解或文件說明使用。

```python
# 這是一行單行註解：下方即將進行圓面積計算
radius = 10  # 也可以把註解寫在代碼的尾端
pi = 3.14159
area = pi * (radius ** 2)

"""
這裡是一整段的多行註解。
你可以在這裡記錄複雜的演算法公式、
修復日誌（Bug Log），或者暫時關閉一大段測試程式碼。
"""
print("計算出的圓面積為:", area)
```

## 6. 探查百寶箱：內建工具 `type()` 與 `dir()`
當專案越來越大，或者引用了別人寫好的未知套件時，你需要兩把強大的放大鏡來檢查變數與模組：

### 🔍 6.1 `type(變數)` — 型態現形鏡
* **功能**：傳回目標物件或變數目前在記憶體中的數據型態類型。不知道這個變數是整數還是字串時，查一下就知道。

### 🔍 6.2 `dir(module_or_object)` — 結構探測雷達
* **功能**：列出該模組（Module）、類別或物件內，**所有可用的函式（Functions）、方法（Methods）與全域屬性變數**的名稱清單。它是你探索新套件最直接的敲門磚。

```python
print("--- 6.1 type() 實戰演練 ---")
x = 42
y = 3.14
z = "Hello Python"

print(f"變數 x ({x}) 的型態是:", type(x))
print(f"變數 y ({y}) 的型態是:", type(y))
print(f"變數 z ('{z}') 的型態是:", type(z))

print("\n--- 6.2 dir() 實戰演練 (以內建 math 模組為例) ---")
import math

# 使用 dir 探測 math 模組裡面有哪些武器可用
math_features = dir(math)

# 由於內容很多，我們印出前 15 個屬性與函式名稱來觀察
print("math 模組內包含的前 15 個成員清單:")
print(math_features[:15])

print("\n💡 測試調用 dir 探查到的成員:")
print("圓周率常數 math.pi =", math.pi)
print("平方根計算 math.sqrt(16) =", math.sqrt(16))
```
