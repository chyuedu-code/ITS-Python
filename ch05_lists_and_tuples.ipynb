# Chapter 5: 串列與元組 (Lists and Tuples)
本章節將深入介紹 Python 中最常用的兩種線性資料結構：**串列 (List)** 與 **元組 (Tuple)**。包含兩者的宣告、索引與切片、常用操作語法、迴圈讀取方法，以及它們之間的轉換。

## 1. 串列基礎與索引切片
串列是一種**可變 (mutable)** 的資料結構，可以容納各種型態的元素。

### 宣告與存取
* **宣告一維串列**：`串列名稱 = [元素1, 元素2, ...]`
* **存取單一元素**：`串列名稱[索引]` (索引從 `0` 開始)
* **範圍切片 (Slicing)**：`串列名稱[起始索引:終止索引:間隔值]`

> ⚠️ **重要錯誤提醒：** 索引不可超出串列範圍，否則執行時將拋出 **`IndexError: list index out of range`** 錯誤！

```python
# 串列宣告與基本存取
fruits = ["蘋果", "香蕉", "芭樂", "葡萄", "西瓜"]
print("原始串列 :", fruits)
print("第一個元素 fruits[0] :", fruits[0])
print("第三個元素 fruits[2] :", fruits[2])

# 切片範例
print("\n--- 切片範例 ---")
print("fruits[1:4]   :", fruits[1:4])   # 索引 1 到 3 (不含 4)
print("fruits[0:5:2] :", fruits[0:5:2]) # 間隔值為 2
print("fruits[::-1]  :", fruits[::-1])  # 技巧：快速反轉串列

# IndexError 測試（取消註解可檢視錯誤）
# print(fruits[10])
```

## 2. 使用迴圈讀取串列
讀取串列主要有兩種 `for` 迴圈的使用方式：
1. **`for 變數 in 串列:`**：直接取出串列中的每個「元素值」（最直覺）。
2. **`for 變數 in range(len(串列)):`**：透過「索引值」間接存取元素（適合需要同時知道位置的場景）。

```python
names = ["小明", "小華", "小紅"]

print("--- 方法 1：直接讀取元素 ---")
for name in names:
    print(name)

print("\n--- 方法 2：透過 range() 索引讀取 ---")
for i in range(len(names)):
    print(f"索引 {i} 的元素是: {names[i]}")
```

## 3. 串列常用操作方法 (Methods)
以下是 Python 串列內建的實用操作：

| 功能 | 語法 | 說明 |
| :--- | :--- | :--- |
| **搜尋索引** | `串列.index(元素)` | 尋找指定元素的索引值 |
| **計算次數** | `串列.count(元素)` | 計算元素在串列中出現的次數 |
| **末端新增** | `串列.append(元素)` | 新增元素在串列的最尾端 |
| **指定插入** | `串列.insert(索引, 元素)` | 在指定索引位置插入元素 |
| **刪除元素** | `串列.remove(元素)` | 刪除串列中第一個匹配的指定元素 |
| **彈出取出** | `串列.pop(索引)` | 取出並同時移除指定位置的元素 |
| **刪除指定索引** | `del 串列[索引]` | 刪除特定位置（支援負索引）的元素 |
| **刪除切片範圍** | `del 串列[起:終:間]` | 批次刪除指定範圍的元素 |
| **內部排序** | `串列.sort()` | 將原本串列直接從小到大排序 (變動原串列) |
| **順序反轉** | `串列.reverse()` | 將目前串列的反序排列 (變動原串列) |
| **保留原樣排序**| `新串列 = sorted(串列)` | 排序並建立新串列，**保留原本串列結構** |

```python
nums = [30, 10, 20, 10, 40]
print("初始數列:", nums)

# 1. index & count
print("20 的索引值為:", nums.index(20))
print("10 出現的次數:", nums.count(10))

# 2. 新增與插入
nums.append(50)
print("append(50) 後:", nums)
nums.insert(1, 15)
print("insert(1, 15) 後:", nums)

# 3. 刪除與彈出
nums.remove(10) # 刪除第一個 10
print("remove(10) 後:", nums)
popped_val = nums.pop(2)
print(f"pop(2) 取出的元素為 {popped_val}, 剩餘串列:", nums)
del nums[0]
print("del nums[0] 後:", nums)

# 4. 排序與反轉
print("\n--- 排序與反轉測試 ---")
my_list = [5, 1, 4, 2]
# 使用 sorted() 保留原串列
new_sorted = sorted(my_list)
print("原串列保留:", my_list)
print("sorted() 新串列:", new_sorted)

# 使用 .sort() 與 .reverse()
my_list.sort()
print(".sort() 原地排序後:", my_list)
my_list.reverse()
print(".reverse() 原地反轉後:", my_list)
```

## 4. 元組 (Tuple) 概念與轉換

### 什麼是元組 (Tuple)？
* **宣告方式**：`元組名稱 = (元素1, 元素2, ...)`
* **核心特性**：元組的**結構與內容不能改變 (immutable)**。你無法對它使用 `append()`、`insert()`、`remove()` 或 `pop()` 等修改語法。

### 為什麼要用元組？
1. **執行速度較快**：系統分配記憶體更具效率。
2. **資料結構較安全**：防止程式意外修改關鍵設定資料（如：資料庫連線參數、GPS 經緯度座標）。

### 串列與元組互相轉換
利用 `list()` 和 `tuple()` 兩個內建建構函式，可以輕鬆轉換資料型態：
* `新串列 = list(元組)`
* `新元組 = tuple(串列)`

```python
# 元組宣告與防錯測試
my_tuple = (10, 20, 30)
print("讀取元組 my_tuple[1] :", my_tuple[1])

# 嘗試修改元組會發生錯誤 (TypeError)
try:
    # my_tuple[0] = 99 # 取消註解會噴錯
    pass
except TypeError as e:
    print("錯誤：元組不能被修改！", e)

# 互相轉換實作
print("\n--- 型態轉換範例 ---")
sample_list = [1, 2, 3]
tupled_list = tuple(sample_list)
print("串列轉元組:", tupled_list, type(tupled_list))

sample_tuple = ("A", "B", "C")
listed_tuple = list(sample_tuple)
listed_tuple.append("D") # 轉成串列後就可以修改了
print("元組轉串列並新增元素:", listed_tuple, type(listed_tuple))
```
