# Chapter 3: 判斷式 (Conditional Statements)
本章節將介紹 Python 中的雙向判斷式、多項判斷式、語法注意事項，以及 Python 特有的 **鏈式比較 (Chained Comparison)** 語法。

## 1. 雙向判斷 (`if...else`)
當條件成立時（`True`）執行程式區塊一；否則（`False`）執行程式區塊二。

### 語法結構
```python
if 條件式:
    程式區塊一
else:
    程式區塊二
```

````python
# 雙向判斷範例：檢查是否及格
score = 75

if score >= 60:
    print("及格！做得好！")
else:
    print("不及格，下次再接再厲！")
````

## 2. 多項判斷 (`if...elif...else`)
當需要檢查多個依序排他的條件時使用。一旦某個條件成立，後續的條件就不會再進行判斷。

### 語法結構
```python
if 條件式一:
    程式區塊一
elif 條件式二:
    程式區塊二
else:
    程式區塊三
```

````python
# 多項判斷範例：成績評等
score = 85

if score >= 90:
    print("等級：A")
elif score >= 80:
    print("等級：B")
elif score >= 70:
    print("等級：C")
else:
    print("等級：D (需要再加油)")
````

> 💡 **小提醒：** 
> 1. **冒號 `:` 不能漏**：不論是 `if`、`elif` 還是 `else` 的尾端，都必須加上冒號 `:`。
> 2. **縮排 (Indentation)**：Python 強制規定屬於該區塊的程式碼必須縮排（通常為 4 個空白鍵），縮排不一致會導致 `IndentationError`。
> 3. **結構邏輯**：多項判斷的條件是由上往下檢查的，前面的條件成立了，後面的就不會看。因此**條件範圍的順序安排非常重要**。

## 3. 鏈式比較在條件判斷中的應用

Python 支援**鏈式比較**（詳見 Chapter 2），讓範圍判斷更加直覺。這個特性在條件判斷中特別實用！

**傳統寫法 vs. 鏈式比較：**
```python
# 傳統寫法（使用 and）
if score >= 60 and score < 80:
    print("及格")

# 鏈式比較（Python 特色）
if 60 <= score < 80:
    print("及格")
```

> 💡 **提示：** 鏈式比較寫法更接近數學表達式，在 ITS Python 考試中是常見的考點！

````python
# 鏈式比較在條件判斷中的實際應用
print("=== 範例 1：成績等級判斷（使用鏈式比較）===")
score = 75

if 90 <= score <= 100:
    print(f"分數 {score}：優秀 (A)")
elif 80 <= score < 90:
    print(f"分數 {score}：良好 (B)")
elif 70 <= score < 80:
    print(f"分數 {score}：中等 (C)")
elif 60 <= score < 70:
    print(f"分數 {score}：及格 (D)")
else:
    print(f"分數 {score}：不及格 (F)")

print("\n=== 範例 2：年齡層分類 ===")
age = 25

if 0 <= age < 13:
    print(f"{age} 歲：兒童")
elif 13 <= age < 20:
    print(f"{age} 歲：青少年")
elif 20 <= age < 60:
    print(f"{age} 歲：成人")
else:
    print(f"{age} 歲：長者")

print("\n=== 範例 3：溫度判斷 ===")
temp = 28

if temp < 10:
    print(f"{temp}°C：寒冷")
elif 10 <= temp < 20:
    print(f"{temp}°C：涼爽")
elif 20 <= temp < 30:
    print(f"{temp}°C：舒適")
else:
    print(f"{temp}°C：炎熱")
````
