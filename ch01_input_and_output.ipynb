本章節將介紹 Python 的基本輸入 `input()`、輸出 `print()`，以及多種強大的字串格式化 (String Formatting) 技巧。

## 1. 基本輸出函式 `print()`
`print()` 是 Python 中最常用的輸出工具，可以同時輸出多個參數，並透過 `sep` 與 `end` 參數自由控制排版。

### 常用選用參數：
* **`sep` (Separator)**：多個輸出值之間的「分隔字元」，預設是一個空白鍵 ``。
* **`end`**：輸出結束後自動附加的「結尾字元」，預設是換行符號 `\n`。

```python
# 基本 print 範例
print("--- 基本輸出 ---")
print("Hello", "Python", 2026)

# 控制 sep 參數
print("\n--- 調整 sep 分隔字元 ---")
print("apple", "banana", "orange", sep="-")
print("2026", "06", "12", sep="/")

# 控制 end 參數
print("\n--- 調整 end 結尾字元（不換行連著印） ---")
print("第一句", end=" *** ")
print("第二句", end="")
print(" (這是同一行)")
```

## 2. 基本輸入函式 `input()`
`input()` 能夠在程式執行時暫停，等待使用者輸入資料。

> ⚠️ **最核心大考點：** `input()` 接收到的所有輸入資料，其**資料型態一律為字串 (str)**！
> 如果需要進行數學運算，必須使用 `int()` 或 `float()` 進行強制型態轉換。

```python
# input 基本範例與型態轉換
name = "小明" # 模擬輸入
print(f"請輸入您的名字: {name}")
print(f"您好，{name}！ 變數型態是: {type(name)}")

print("\n--- 數字計算必須轉換型態 ---")
str_age = "18" # 模擬輸入的字串數字
print(f"請輸入您的年齡: {str_age}")

# 錯誤示範（取消註解會噴錯：不能將字串與數字相加）
# next_year = str_age + 1 

# 正確示範：強制轉成整數 (int)
int_age = int(str_age)
print(f"明年您的年齡將是: {int_age + 1} 歲")
```

## 3. 字串格式化參數 (String Formatting)
在輸出複雜的字串與變數組合時，硬用 `+` 拼接代碼會非常凌亂。Python 提供了三種主流且好用的格式化方法：

### ① 最經典：`.format()` 語法
使用大括號 `{}` 作為變數的佔位符，並在字串後面串接 `.format(變數1, 變數2)`。
* 可使用預設依序填入：`"{} {}".format(a, b)`
* 可使用索引編號填入：`"{1} {0}".format(a, b)`

### ② 最推薦、最新穎：f-string (字串字面值格式化)
自 Python 3.6 起支援。只需在字串最前端加上 `f` 或 `F`，就能直接在大括號 `{}` 內部寫入變數名稱或運算式，直覺、易讀且執行效能最高！

### ③ 數值格式微調 (置中、靠左與小數點控制)
不論在 `.format()` 或 `f-string` 中，都可以透過 `{變數:格式}` 機制控制排版：
* `:.2f`：四捨五入保留小數點後兩位。
* `:>10`：靠右對齊，總寬度佔 10 格。
* `:<10`：靠左對齊，總寬度佔 10 格。
* `:^10`：置中對齊，總寬度佔 10 格。

```python
item = "手機"
price = 24500.567

print("--- 1. .format() 方法 ---")
print("品名：{}，價格：{}".format(item, price))
print("指定順序 -> 價格：{1}，品名：{0}".format(item, price))

print("\n--- 2. f-string 方法 (最推薦) ---")
print(f"品名：{item}，價格：{price}")
print(f"價格乘兩倍是：{price * 2}")

print("\n--- 3. 數值與排版格式微調 ---")
# 小數點控制
print(f"價格(保留兩位小數): {price:.2f}")

# 排版對齊控制
print(f"靠左對齊: |{item:<10}|")
print(f"靠右對齊: |{item:>10}|")
print(f"置中對齊: |{item:^10}|")
```
