# Chapter 6: 集合與字典 (Sets and Dictionaries)
本章節將帶您完整掌握 Python 的兩種雜湊（Hash）資料結構：**集合 (Set)** 與 **字典 (Dictionary)**。
內容包含兩者的基礎特徵、CRUD（增刪查改）基本操作、走訪 (Iteration) 技巧，以及常見的特性對比。

## 1. 集合 (Set) 的 CRUD 與 走訪
集合是**無序 (unordered)**、**元素不重複 (unique)** 的容器。內部元素必須是不可變的（hashable）。

### CRUD 操作語法：
* **Create (建立)**：`my_set = {1, 2, 3}` 或 `empty_set = set()` (注意：`{}` 預設代表空字典)。
* **Read (讀取/檢查)**：無法用索引讀取，但可以用 `in` 運算子檢查元素是否存在。
* **Update (新增/修改)**：`.add(元素)` 新增單一元素；`.update(可迭代物件)` 批次新增多個元素。
* **Delete (刪除)**：
  * `.remove(元素)`: 刪除指定元素，若不存在會**報錯**。
  * `.discard(元素)`: 刪除指定元素，若不存在**不會報錯**（安全）。
  * `.pop()`: 隨機彈出並刪除一個元素。
  * `.clear()`: 清空集合。

````python
# 集合 CRUD 實作
print("--- 1. 建立與檢查 (Create & Read) ---")
colors = {"red", "green", "blue"}
print("初始集合:", colors)
print("檢查 'red' 是否在集合中:", "red" in colors)

print("\n--- 2. 新增與更新 (Update) ---")
colors.add("yellow")
colors.add("red") # 重複加入，會被自動忽略
print("add() 後:", colors)
colors.update(["black", "white"])
print("update() 後:", colors)

print("\n--- 3. 刪除操作 (Delete) ---")
colors.remove("black")
print("remove('black') 後:", colors)
colors.discard("purple") # 不存在的元素，安全不報錯
print("discard('purple') 後:", colors)

print("\n--- 4. 集合走訪 (Iteration) ---")
for color in colors:
    print(f"走訪元素: {color}")
````

## 2. 字典 (Dictionary) 的 CRUD 與 走訪
字典是由 **鍵-值對 (key-value pairs)** 組成的對映結構。`Key` 必須唯一且不可變，`Value` 則可以重複且為任意型態。

### CRUD 操作語法：
* **Create (建立)**：`my_dict = {"name": "小明", "age": 18}`
* **Read (讀取)**：
  * `dict[key]`: 直覺讀取，若 Key 不存在會**拋出 KeyError**。
  * `dict.get(key, 預設值)`: 安全讀取，若不存在則回傳預設值（不報錯）。
* **Update (新增/修改)**：`dict[key] = value` (Key 存在即修改，不存在即新增)。或使用 `.update(新字典)`。
* **Delete (刪除)**：
  * `del dict[key]`: 刪除指定鍵值對。
  * `.pop(key)`: 取出並移除指定鍵值對。

````python
# 字典 CRUD 實作
print("--- 1. 建立與讀取 (Create & Read) ---")
user = {"name": "小明", "age": 18}
print("user['name'] :", user["name"])
print("get() 安全讀取不存在的 key:", user.get("gender", "未知"))

print("\n--- 2. 新增與修改 (Update) ---")
user["age"] = 19       # 修改既存的 Key
user["city"] = "台北"  # 新增不存在的 Key
print("更新後的字典:", user)

print("\n--- 3. 刪除操作 (Delete) ---")
removed_val = user.pop("age")
print(f"pop('age') 取出的值為 {removed_val}, 剩餘字典:", user)
del user["city"]
print("del user['city'] 後:", user)

print("\n--- 4. 字典的三種走訪方法 (Iteration) ---")
info = {"id": 101, "item": "筆電", "status": "出貨中"}
# 方法 A：唯獨走訪鍵 (Keys)
print("A. 走訪 Keys:")
for k in info.keys():
    print(f"  Key: {k}")

# 方法 B：唯獨走訪值 (Values)
print("B. 走訪 Values:")
for v in info.values():
    print(f"  Value: {v}")

# 方法 C：同時走訪鍵與值 (Items) - 最常用！
print("C. 同時走訪 Items:")
for k, v in info.items():
    print(f"  鍵: {k} -> 值: {v}")
````

## 3. Comprehension（推導式）：優雅簡潔的資料生成技巧
Comprehension 是 Python 獨特且強大的語法糖，能讓我們用**一行程式碼**就生成列表、字典或集合，大幅簡化傳統迴圈的寫法。

### 📝 三大推導式語法：

#### 1. 列表推導式 (List Comprehension)
```python
[expression for item in iterable if condition]
```

#### 2. 字典推導式 (Dict Comprehension)
```python
{key_expr: value_expr for item in iterable if condition}
```

#### 3. 集合推導式 (Set Comprehension)
```python
{expression for item in iterable if condition}
```

### 💡 核心優勢：
* **簡潔易讀**：一行程式碼取代多行迴圈
* **效能較佳**：內部優化，執行速度通常比傳統迴圈快
* **函式式風格**：符合 Python 的優雅哲學

````python
print("--- 1. 列表推導式 (List Comprehension) ---")
# 傳統寫法：產生 1-10 的平方數
squares_old = []
for x in range(1, 11):
    squares_old.append(x ** 2)
print("傳統迴圈結果:", squares_old)

# 推導式寫法：一行搞定！
squares = [x ** 2 for x in range(1, 11)]
print("推導式結果:", squares)

# 帶條件的推導式：只取偶數的平方
even_squares = [x ** 2 for x in range(1, 11) if x % 2 == 0]
print("偶數平方:", even_squares)

# 字串處理：轉大寫
fruits = ['apple', 'banana', 'cherry']
upper_fruits = [fruit.upper() for fruit in fruits]
print("轉大寫:", upper_fruits)

print("\n--- 2. 字典推導式 (Dict Comprehension) ---")
# 產生數字與其平方的對映
square_dict = {x: x ** 2 for x in range(1, 6)}
print("數字平方字典:", square_dict)

# 從兩個列表建立字典
names = ['Alice', 'Bob', 'Charlie']
ages = [25, 30, 35]
name_age_dict = {name: age for name, age in zip(names, ages)}
print("姓名年齡對映:", name_age_dict)

# 字典鍵值互換
original = {'a': 1, 'b': 2, 'c': 3}
swapped = {v: k for k, v in original.items()}
print("鍵值互換:", swapped)

# 帶條件的字典推導式：只保留偶數值
filtered_dict = {k: v for k, v in square_dict.items() if v % 2 == 0}
print("只保留偶數值:", filtered_dict)

print("\n--- 3. 集合推導式 (Set Comprehension) ---")
# 產生 1-10 的平方數集合（自動去重）
square_set = {x ** 2 for x in range(1, 11)}
print("平方數集合:", square_set)

# 從字串中取出唯一的字元
text = "hello world"
unique_chars = {char for char in text if char != ' '}
print("唯一字元:", unique_chars)

# 取出列表中的唯一長度
words = ['apple', 'bat', 'car', 'dog', 'elephant']
lengths = {len(word) for word in words}
print("單字長度集合:", lengths)

print("\n--- 4. 巢狀推導式 (Nested Comprehension) ---")
# 產生九九乘法表
multiplication_table = [[i * j for j in range(1, 10)] for i in range(1, 10)]
print("九九乘法表（前3行）:")
for row in multiplication_table[:3]:
    print(row)

# 扁平化二維列表
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
flattened = [num for row in matrix for num in row]
print("扁平化結果:", flattened)

print("\n--- 5. 實際應用範例 ---")
# 範例A：過濾並轉換資料
scores = {'Alice': 85, 'Bob': 92, 'Charlie': 78, 'David': 95}
high_scorers = {name: score for name, score in scores.items() if score >= 90}
print("高分者 (>=90):", high_scorers)

# 範例B：資料清理
raw_data = ['  apple  ', '  Banana', 'CHERRY  ']
clean_data = [item.strip().lower() for item in raw_data]
print("清理後的資料:", clean_data)

# 範例C：快速建立查找表
products = [('P001', 'Laptop'), ('P002', 'Mouse'), ('P003', 'Keyboard')]
product_lookup = {code: name for code, name in products}
print("產品查找表:", product_lookup)
````

## 4. 資料結構綜合對比 (Comparison)
為了方便釐清 Python 常用四大容器的差異，以下提供完整的特徵對照表：

| 資料結構 | 符號範例 | 有序/無序 | 允許重複 | 元素存取方式 | 可變性 (Mutable) |
| :--- | :--- | :--- | :--- | :--- | :--- |
| **串列 List** | `[1, 2, 3]` | 有序 (Ordered) | 允許 | 透過整數索引 `list[0]` | **可變** |
| **元組 Tuple** | `(1, 2, 3)` | 有序 (Ordered) | 允許 | 透過整數索引 `tuple[0]` | **不可變** |
| **集合 Set** | `{1, 2, 3}` | 無序 (Unordered) | **不允許** | 無索引，僅能用 `in` 或走訪 | **可變** (但元素須不可變) |
| **字典 Dict** | `{"a": 1}` | 有序\* (Python 3.7+) | Key不重 / Value可重 | 透過自訂的鍵 `dict["a"]` | **可變** |

\* *註：自 Python 3.7 起，字典內部實作正式保證了 Key 的「插入順序 (Insertion Order)」，但在概念上它仍屬於依賴 Key-Value 對映（Mapping）而非依賴序列索引（Index）存取的結構。*
